```python
import matplotlib.pyplot as plt

def solicitar_dados():
    # Solicitar o número de pontos experimentais
    n = int(input("Quantos pontos experimentais você tem? "))

    # Inicializar listas para armazenar os dados
    I_exp = []
    desvio_padrao = []

    # Solicitar os valores experimentais e de desvio padrão
    for i in range(n):
        valor_exp = float(input(f"Insira o valor experimental {i+1}: "))
        desvio = float(input(f"Insira o desvio padrão para o ponto {i+1}: "))
        I_exp.append(valor_exp)
        desvio_padrao.append(desvio)

    return I_exp, desvio_padrao

def plotar_grafico(I_exp, desvio_padrao):
    # Calcular valores teóricos
    I_teo = [I_exp[0]*(2*i+1) for i in range(len(I_exp))]

    # Ajuste do tamanho do gráfico
    plt.figure(figsize=(10, 6))

    # Plotando os gráficos com escala logarítmica no eixo Y
    plt.errorbar(range(len(I_exp)), I_exp, yerr=desvio_padrao, fmt='o', label='Experimental', capsize=5)
    plt.plot(range(len(I_exp)), I_teo, 's--', label='Teórico')

    # Aplicando a escala logarítmica no eixo Y
    plt.yscale('log')

    # Adicionando grade
    plt.grid(True)

    # Título e rótulos
    plt.title("Deslocamento Lateral")
    plt.xlabel("Ordem $m$")
    plt.ylabel("Intensidade (Escala Log)")

    # Legenda
    plt.legend()

    # Ajuste dos ticks do eixo X
    plt.xticks(range(len(I_exp)))

    # Mostrando o gráfico
    plt.show()

# Solicitar dados do usuário
I_exp, desvio_padrao = solicitar_dados()

# Plotar gráfico
plotar_grafico(I_exp, desvio_padrao)
```
