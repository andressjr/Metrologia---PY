```python

def solicitar_dados():
    n = int(input("Quantos pontos experimentais você tem? "))
    I_exp = []
    for i in range(n):
        valor_exp = float(input(f"Insira o valor experimental {i+1}: "))
        I_exp.append(valor_exp)
    return I_exp

def plotar_grafico(I_exp):
    I_teo = [I_exp[0]*(2*i+1) for i in range(len(I_exp))]

    # Ajuste do tamanho do gráfico
    plt.figure(figsize=(12, 6))

    # Gráfico em escala logarítmica
    plt.subplot(1, 2, 1)
    plt.plot(range(len(I_exp)), I_exp, 'o-', label='Experimental')
    plt.plot(range(len(I_exp)), I_teo, 's--', label='Teórico')
    plt.yscale('log')
    plt.title("Deslocamento Lateral - Logarítmico")
    plt.xlabel("Ordem $m$")
    plt.ylabel("Intensidade (Escala Log)")
    plt.grid(True)
    plt.legend()

    # Gráfico em escala linear
    plt.subplot(1, 2, 2)
    plt.plot(range(len(I_exp)), I_exp, 'o-', label='Experimental')
    plt.plot(range(len(I_exp)), I_teo, 's--', label='Teórico')
    plt.title("Deslocamento Lateral - Linear")
    plt.xlabel("Ordem $m$")
    plt.ylabel("Intensidade")
    plt.grid(True)
    plt.legend()

    # Ajuste dos ticks do eixo X para ambos os gráficos
    plt.xticks(range(len(I_exp)))

    # Mostrando o gráfico
    plt.tight_layout()
    plt.show()

# Solicitar dados do usuário
I_exp = solicitar_dados()

# Plotar gráfico
plotar_grafico(I_exp)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
I_3500 = [22.88, 63, 108, 135, 182, 223, 268, 312, 344, 388, 431, 481, 522, 568, 613]
I_2000 = [370, 409, 435, 463, 491, 537]

# Combinação dos dados de I_3500 e I_2000
I_exp = [I_3500[i] if i < len(I_3500) else I_2000[i - len(I_3500)] * 3500 / 2000 for i in range(len(I_3500) + len(I_2000))]

# Realize o ajuste linear com apenas um parâmetro livre 'a'
a = np.polyfit(range(len(I_exp)), I_exp, 1)[0]

# Calcular os valores previstos para a reta ajustada
y_pred = a * np.arange(len(I_exp)) + (a / 2)

# Plotar os dados experimentais e a reta ajustada
plt.figure(figsize=(10, 5))
plt.scatter(range(len(I_exp)), I_exp, label='Dados Experimentais')
plt.plot(range(len(I_exp)), y_pred, color='red', linestyle='--', label=f'Ajuste Linear: y = {a:.2f}x + {a/2:.2f}')
plt.xlabel('Índice')
plt.ylabel('Valores de I_exp')
plt.legend()
plt.grid(True)
plt.show()

# Coeficiente 'a' calculado
print(f'Coeficiente "a": {a:.2f}')
```
