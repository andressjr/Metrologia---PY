```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Dados
I_exp = [30.49, 80.25, 136.2, 190.1, 240.8, 302, 349.5, 460.7, 492.2, 577.1, 591.3, 634.2, 698.5, 745.8, 837.2, 873.5, 890.4, 972, 1044, 1098]
I_teo = [I_exp[0]*(2*i+1) for i in range(len(I_exp))]

# Ajuste do tamanho do grfico
plt.figure(figsize=(10, 6))

# Plotando os grficos com escala logartmica no eixo Y
plt.plot(range(len(I_exp)), I_exp, 'o--', label='Experimental')
plt.plot(range(len(I_exp)), I_teo, 's--', label='Terico')

# Aplicando a escala logartmica no eixo Y
plt.yscale('log')

# Adicionando grade
plt.grid(True)

# Ttulo e rtulos
plt.title("Deslocamento Lateral")
plt.xlabel("Ordem $m$")
plt.ylabel("Intensidade (Escala Log)")

# Legenda
plt.legend()

# Ajuste dos ticks do eixo X
plt.xticks(range(len(I_exp)))

# Mostrando o grfico
plt.show()
```

```python

def solicitar_dados():
    n = int(input("Quantos pontos experimentais voc tem? "))
    I_exp = []
    for i in range(n):
        valor_exp = float(input(f"Insira o valor experimental {i+1}: "))
        I_exp.append(valor_exp)
    return I_exp

def plotar_grafico(I_exp):
    I_teo = [I_exp[0]*(2*i+1) for i in range(len(I_exp))]

    # Ajuste do tamanho do grfico
    plt.figure(figsize=(12, 6))

    # Grfico em escala logartmica
    plt.subplot(1, 2, 1)
    plt.plot(range(len(I_exp)), I_exp, 'o-', label='Experimental')
    plt.plot(range(len(I_exp)), I_teo, 's--', label='Terico')
    plt.yscale('log')
    plt.title("Deslocamento Lateral - Logartmico")
    plt.xlabel("Ordem $m$")
    plt.ylabel("Intensidade (Escala Log)")
    plt.grid(True)
    plt.legend()

    # Grfico em escala linear
    plt.subplot(1, 2, 2)
    plt.plot(range(len(I_exp)), I_exp, 'o-', label='Experimental')
    plt.plot(range(len(I_exp)), I_teo, 's--', label='Terico')
    plt.title("Deslocamento Lateral - Linear")
    plt.xlabel("Ordem $m$")
    plt.ylabel("Intensidade")
    plt.grid(True)
    plt.legend()

    # Ajuste dos ticks do eixo X para ambos os grficos
    plt.xticks(range(len(I_exp)))

    # Mostrando o grfico
    plt.tight_layout()
    plt.show()

# Solicitar dados do usurio
I_exp = solicitar_dados()

# Plotar grfico
plotar_grafico(I_exp)
```
